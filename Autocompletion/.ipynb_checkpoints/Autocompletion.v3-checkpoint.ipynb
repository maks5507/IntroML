```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import sklearn
```

```python
df = pd.read_csv("train.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
firstWordCounter = {}
secondWordCounter = {}
pairsWordCounter = {}
for pair in df.Prediction:
    if pair.decode('utf8') in pairsWordCounter:
        pairsWordCounter[pair.decode('utf8')] += 1
    else:
        pairsWordCounter[pair.decode('utf8')] = 1
    temp = pair.split(' ')
    first = temp[0].decode('utf8')
    second = temp[1].decode('utf8')
    if first in firstWordCounter:
        firstWordCounter[first] += 1
    else:
        firstWordCounter[first] = 1
    if second in secondWordCounter:
        secondWordCounter[second] += 1
    else:
        secondWordCounter[second] = 1
```

```python
feminine = set()
male = set()
neuter = set()
multiple = set()
none = set()
for p in secondWordCounter.keys():
    temp = p[len(p) - 2:]
    if temp == u'ла' or temp == u'ая':
        feminine.add(p)
    elif temp == u'ло' or temp == u'ое':
        neuter.add(p)
    elif temp == u'ли' or temp == u'ые':
        multiple.add(p)
    elif temp == u'ый':
        male.add(p)
    else:
        none.add(p)
        
```

```python
print len(feminine)
print len(male)
print len(neuter)
print len(multiple)
print len(none)
```

```python
print(len(firstWordCounter))
print(len(secondWordCounter))
print(len(pairsWordCounter))
```

```python
proba = {}
omegaAB = len(pairsWordCounter)
omegaB = len(firstWordCounter)
for i, pair in enumerate(pairsWordCounter.keys()):
    word = pair.split(' ')
    pab = 1.0 * pairsWordCounter[pair] / omegaAB
    pb = 1.0 * firstWordCounter[word[0]] / omegaB
    proba[pair] = 1.0 * pab / pb
```

```python
prefix = {}
for word in secondWordCounter.keys():
    pref = word[:2]
    if not pref in prefix:
        prefix[pref] = []
    prefix[pref].append(word)
```

```python
dft = pd.read_csv('test.csv')
```

```python
misses = []
answToSubmit = []
for i, pair in enumerate(dft.Sample):
    if i % 100000 == 0:
        print i
    pair = pair.decode('utf8')
    word = pair.split(' ')
    ls = len(word[1])
    temp = word[1][:2]
    test = []
    if not temp in prefix:
        answToSubmit.append('#')
        continue
    for p in prefix[temp]:
        if p[:ls] == word[1]:
            test.append(p)
    maxProba = 0.0
    maxFreq = 0.0
    answP = ""
    answF = ""
    for p in test:
        ab = (' ').join([word[0], p])
        if maxFreq < secondWordCounter[p]:
            maxFreq, answF = secondWordCounter[p], ab
        if not ab in proba:
            continue
        if maxProba < proba[ab]:
            maxProba, answP = proba[ab], ab
    if maxProba == 0.0:
        answ = answF
    else:
        answ = answP
    if answ == '':
        answ = '#'
    answToSubmit.append(answ)
```
