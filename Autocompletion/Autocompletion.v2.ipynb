```python
firstWordCounter = {}
secondWordCounter = {}
pairsWordCounter = {}
for pair in df.Prediction:
    if pair.decode('utf8') in pairsWordCounter:
        pairsWordCounter[pair.decode('utf8')] += 1
    else:
        pairsWordCounter[pair.decode('utf8')] = 1
    temp = pair.split(' ')
    first = temp[0].decode('utf8')
    second = temp[1].decode('utf8')
    if first in firstWordCounter:
        firstWordCounter[first] += 1
    else:
        firstWordCounter[first] = 1
    if second in secondWordCounter:
        secondWordCounter[second] += 1
    else:
        secondWordCounter[second] = 1
```

```python
roots = {}
for i in [2, 1, 0]:
    for p in pairsWordCounter.keys():
        words = p.split(' ')
        f = words[0]
        s = words[1]
        ft = f[:len(f) - i]
        st = s[:len(s) - i]
        if not ft in roots:
            roots[ft] = set()
        if not st in roots:
            roots[st] = set()
        roots[ft].add(f)
        roots[st].add(s)
```

```python
for p in roots.keys():
    if len(roots[p]) == 1:
        roots.pop(p)
```

```python
transformed = {}
transformedPairsWordCounter = {}
for p in pairsWordCounter.keys():
    words = p.split(' ')
    f = words[0]
    s = words[1]
    transformed[f] = f
    transformed[s] = s
    for i in range(0, 3):
        ft = f[:len(f) - i]
        st = s[:len(s) - i]
        if ft in roots:
            transformed[f] = ft
        if st in roots:
            transformed[s] = st

for p in pairsWordCounter.keys():
    words = p.split(' ')
    f = words[0]
    s = words[1]
    tr = (' ').join([f, transformed[s]])
    if not tr in transformedPairsWordCounter:
        transformedPairsWordCounter[tr] = 0
    transformedPairsWordCounter[tr] += 1
```

```python
transformedFirstWordCounter = {}
transformedSecondWordCounter = {}
for p in transformedPairsWordCounter.keys():
    words = p.split(' ')
    f = words[0]
    s = words[1]
    if not f in transformedFirstWordCounter:
        transformedFirstWordCounter[f] = 0
    if not s in transformedSecondWordCounter:
        transformedSecondWordCounter[s] = 0
    transformedFirstWordCounter[f] += 1
    transformedSecondWordCounter[s] += 1
```

```python
proba = {}
omegaAB = len(transformedPairsWordCounter)
omegaB = len(firstWordCounter)
for i, pair in enumerate(transformedPairsWordCounter.keys()):
    word = pair.split(' ')
    pab = 1.0 * transformedPairsWordCounter[pair] / omegaAB
    pb = 1.0 * firstWordCounter
    [word[0]] / omegaB
    proba[pair] = 1.0 * pab / pb
```

```python
prefix = {}
for word in transformedSecondWordCounter.keys():
    pref = word[:2]
    if not pref in prefix:
        prefix[pref] = []
    prefix[pref].append(word)
```

```python
misses = []
answToSubmit = []
for i, pair in enumerate(dft.Sample[:1000]):
    if i % 100000 == 0:
        print i
    pair = pair.decode('utf8')
    word = pair.split(' ')
    ls = len(word[1])
    temp = word[1][:2]
    test = []
    if not temp in prefix:
        answToSubmit.append('#')
        continue
    for p in prefix[temp]:
        if p[:ls] == word[1]:
            test.append(p)
    maxProba = 0.0
    maxFreq = 0.0
    answP = ""
    answF = ""
    for p in test:
        ab = (' ').join([word[0], p])
        if maxFreq < transformedSecondWordCounter[p]:
            maxFreq, answF = transformedSecondWordCounter[p], ab
        if not ab in proba:
            continue
        if maxProba < proba[ab]:
            maxProba, answP = proba[ab], ab
    if maxProba == 0.0:
        answ = answF
    else:
        answ = answP
    if answ == '':
        answ = pair
    answToSubmit.append(answ)
```
